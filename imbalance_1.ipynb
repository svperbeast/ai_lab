```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

from utils import SMOTE

dt = pd.read_csv('nonsep_3clusters_high_noise.csv')
X = np.array(dt.loc[:, dt.columns != 'y'])
y = np.array(dt.y)
Counter(y)
```

```python
smt = SMOTE(X,y,K=5,dup_size=1)
smt_X = np.vstack((X,smt))
smt_y = np.concatenate((y,np.ones(smt.shape[0])))
```

```python
plt.plot(X[y==0,0],X[y==0,1],'.',c='black')
plt.plot(X[y==1,0],X[y==1,1],'.',c='red')
plt.show()

plt.plot(smt_X[smt_y==0,0],smt_X[smt_y==0,1],'.',c='black')
plt.plot(smt_X[smt_y==1,0],smt_X[smt_y==1,1],'.',c='red')
plt.show()
```
